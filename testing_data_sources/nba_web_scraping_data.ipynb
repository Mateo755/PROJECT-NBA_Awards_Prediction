Webscraping with BeautifulSoup

```python
import requests
from bs4 import BeautifulSoup, Comment
import pandas as pd

url = "https://www.basketball-reference.com/leagues/NBA_2024_standings.html"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# Szukamy ukrytych tabel w komentarzach
comments = soup.find_all(string=lambda text: isinstance(text, Comment))

# Przeszukujemy komentarze pod kątem tabel HTML
for c in comments:
    if 'table' in c:
        comment_soup = BeautifulSoup(c, "html.parser")
        tables = comment_soup.find_all("table")
        for t in tables:
            print(t.get("id"))  # pokaże ID każdej tabeli
```

```python
# Wyciągnięcie pełnej tabeli expanded_standings
for c in comments:
    if 'expanded_standings' in c:
        comment_soup = BeautifulSoup(c, "html.parser")
        table = comment_soup.find("table", {"id": "expanded_standings"})
        df_expanded = pd.read_html(str(table))[0]
        break  # nie musimy sprawdzać dalej

df_expanded.head()
```
